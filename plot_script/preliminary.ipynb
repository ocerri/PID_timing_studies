```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mass = [0.1, 1,10,100, 1000]
p = np.arange(0.001, 100, 0.1)
b = []
for m in mass:
    b.append(1./np.sqrt(1+m/p**2))
```

```python
for i, m in enumerate(mass):
    plt.plot(p,b[i], label='m = {} GeV'.format(m))
plt.plot([55],[0.92],'*k',markersize=10, label='Meas. track param.')
plt.legend(loc='best')
plt.ylabel(r'$\beta$')
plt.xlabel('p [GeV]')
plt.grid()
plt.savefig('beta_vs_p')
```
